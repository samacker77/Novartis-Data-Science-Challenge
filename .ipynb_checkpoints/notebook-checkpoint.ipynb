<h3 style="text-align:center">Importing libraries</h3>

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
def load_data(name):
    return pd.read_csv('Dataset/'+name+'.csv')
```

```python
train_df = load_data('Train')
test_df = load_data('Test')
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
train_df = train_df.set_index(train_df.INCIDENT_ID)
test_df = test_df.set_index(test_df.INCIDENT_ID)
```

```python
train_df.shape
```

```python
test_df.shape
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
train_df.drop('INCIDENT_ID',axis=1,inplace=True)
test_df.drop('INCIDENT_ID',axis=1,inplace=True)
```

```python
train_df.shape
```

```python
test_df.shape
```

```python
train_df.X_12 = train_df.X_12.fillna(train_df.X_12.mode()[0])
test_df.X_12 = test_df.X_12.fillna(test_df.X_12.mode()[0])
```

```python
print(train_df.isna().any())
print(test_df.isna().any())
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
def get_day(instance):
    return instance[:2]
def get_month(instance):
    return instance[3:6]
def get_year(instance):
    return instance[7:]
```

```python
train_df['DATE_DAY'] = train_df.DATE.apply(get_day)
test_df['DATE_DAY'] = test_df.DATE.apply(get_day)

train_df['DATE_MONTH'] = train_df.DATE.apply(get_month)
test_df['DATE_MONTH'] = test_df.DATE.apply(get_month)

train_df['DATE_YEAR'] = train_df.DATE.apply(get_year)
test_df['DATE_YEAR'] = test_df.DATE.apply(get_year)
```

```python
train_df.drop('DATE',axis=1,inplace=True)
test_df.drop('DATE',axis=1,inplace=True)

test_df.head()
```

```python
train_df['DATE_MONTH'] = pd.to_datetime(train_df.DATE_MONTH, format='%b').dt.month
test_df['DATE_MONTH'] = pd.to_datetime(test_df.DATE_MONTH, format='%b').dt.month
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
train_df_labels = train_df.MULTIPLE_OFFENSE
train_df_features = train_df.drop('MULTIPLE_OFFENSE',axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
train_data = scaler.fit_transform(train_df_features)
test_data = scaler.transform(test_df)
```

```python
train_X = pd.DataFrame(train_data,columns=train_df_features.columns,index=train_df.index)
test_X = pd.DataFrame(test_data,columns=test_df.columns,index=test_df.index)
```

```python
train_X.head()
```

```python
test_X.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import cross_val_score
```

```python
lr = LogisticRegression()
```

```python
cross_val_score(lr,train_X,train_df_labels,cv=5)
```

```python
lr.fit(train_X,train_df_labels)
```

```python
test_X.head()
```

```python
preds = lr.predict(test_X)
```

```python
submission_df = pd.DataFrame({'MULTIPLE_OFFENSE':preds},index=test_X.index)
```

```python
submission_df.head()
```

```python
submission_df.to_csv('submission.csv')
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier()
```

```python
rf.fit(train_X,train_df_labels)
```

```python
rf_preds = rf.predict(test_X)
```

```python
submission_rf = pd.DataFrame({'MULTIPLE_OFFENSE':rf_preds},index=test_X.index)
```

```python
submission_rf.to_csv('rfsubmission.csv')
```
